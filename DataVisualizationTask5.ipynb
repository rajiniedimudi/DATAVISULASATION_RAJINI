```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("Housing.csv")

# Custom color scales
custom_scale_treemap = [
    (0.0, "pink"),
    (0.5, "skyblue"),
    (1.0, "tan")
]

custom_scale_sunburst = [
    (0.0, "orange"),
    (0.5, "beige"),
    (1.0, "tan")
]

# ---------------- TreeMap Visualization ----------------
fig_tree = px.treemap(
    df,
    path=['furnishingstatus', 'bedrooms', 'parking'],  # Hierarchy
    values='price',
    color='price',
    color_continuous_scale=custom_scale_treemap,
    title="TreeMap: Housing Prices by Furnishing, Bedrooms, and Parking"
)
fig_tree.show()

# ---------------- Sunburst Visualization ----------------
fig_sunburst = px.sunburst(
    df,
    path=['furnishingstatus', 'bedrooms', 'parking'],
    values='price',
    color='price',
    color_continuous_scale=custom_scale_sunburst,
    title="Sunburst: Housing Prices by Furnishing, Bedrooms, and Parking"
)
fig_sunburst.show()
```
